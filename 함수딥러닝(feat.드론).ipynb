```python
import tensorflow as tf
import numpy as np


xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0])
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10., 13.])

# 학습이 안된 신경망
model = tf.keras.Sequential([
                             tf.keras.layers.InputLayer(input_shape=(1,)), #입력 x
                             tf.keras.layers.Dense(1)  #(1)은 출력 y의 개수
])

model.compile(optimizer='sgd', loss='mean_squared_error')

# 학습 횟수는 5회(공부) 횟수가 많아질수록 정확한 답을 도출함(현 예제에서는 31에 가까운)
model.fit(xs, ys, epochs=5)

# 시험 보는 것
p = model.predict([10.0])

print('p:', p)
```
